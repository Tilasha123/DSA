```python
9)A music streaming application stores songs in the order they are added.
Requirements:
• Add a new song at the end.
• Display the playlist.
• Delete a song by title.
Tasks:
• Explain why Singly Linked List is suitable.
• Write Python code for the required operations.
• Analyze the time complexity of each operation.
```

```python
class Node:
    def __init__(self, title):
        self.title = title
        self.next = None


class Playlist:
    def __init__(self):
        self.head = None

    # Add a new song
    def add_song(self, title):
        new_node = Node(title)

        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new_node

        print(f"{title} added to playlist")

    # Display the entire playlist
    def display(self):
        if self.head is None:
            print("Playlist is empty.")
            return

        current = self.head
        position = 1

        while current is not None:
            print(f"{position}. {current.title}")
            current = current.next
            position += 1

    # Delete a song by title
    def delete_song(self, title):
        if self.head is None:
            print("Playlist is empty.")
            return

        if self.head.title == title:
            self.head = self.head.next
            print("Item deleted")
            return

        current = self.head

        while current.next is not None:
            if current.next.title == title:
                current.next = current.next.next
                print("Item deleted")
                return
            current = current.next

        print("Item not found")
```

```python
10) A web browser maintains browsing history where users can:
• Move to the next page.
• Move back to the previous page.
Tasks:

• Explain why Doubly Linked List is appropriate.
• Draw the structure.
• Implement navigation operations in Python.
```

```python
class PageNode:
    def __init__(self, url):
        self.url = url
        self.prev = None
        self.next = None

class BrowserHistory:
    def __init__(self, homepage):
        self.current = PageNode(homepage)
        print(f"  Opened: {homepage}")

    # Visit a new URL 
    def visit(self, url):
        new_page = PageNode(url)
        new_page.prev = self.current
        self.current.next = new_page   # discard any forward history
        self.current = new_page
        print(f"  Visited: {url}")

    # Go back one page 
    def back(self):
        if self.current.prev:
            self.current = self.current.prev
            print(f"Back to: {self.current.url}")
        else:
            print("  No previous page.")

    # Go forward one page
    def forward(self):
        if self.current.next:
            self.current = self.current.next
            print(f"Forward to: {self.current.url}")
        else:
            print("  No next page.")

    def current_page(self):
        print(f"  Current: {self.current.url}")
```

```python
browser = BrowserHistory("google.com")
```

```python
browser.visit("youtube.com")
```

```python
browser.current_page()
```

```python
11) A multiplayer game rotates turns among players continuously.
Tasks:
• Explain why Circular Singly Linked List is suitable.
• Draw the structure for 5 players.
• Write Python code to simulate player turns.
```

```python
class PlayerNode:
    def __init__(self, name):
        self.name = name
        self.next = None


class GameRotation:
    def __init__(self):
        self.head = None
        self.current = None

    # Add player to the circle
    def add_player(self, name):
        new_node = PlayerNode(name)

        if self.head is None:
            self.head = new_node
            new_node.next = self.head
            self.current = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next

            temp.next = new_node
            new_node.next = self.head

    # Simulate multiple rounds
    def simulate(self, rounds):
        if self.head is None:
            print("No players in the game.")
            return

        print(f"\nSimulating {rounds} turns:")
        for turn in range(rounds):
            print(f"Turn {turn + 1}: {self.current.name}")
            self.current = self.current.next
```

```python
game = GameRotation()
for name in ["Ankur", "Aashish", "Kishor", "Abijeet", "Prabin"]:
    game.add_player(name)
game.simulate(8)   # 8 turns across 5 players shows looping
```

```python
12)A music player supports:
• Next Song
• Previous Song
• Continuous looping
Tasks:
• Explain why Doubly Circular Linked List is the best choice.
• Draw the structure.
• Implement next and previous navigation in Python.
```

```python
class SongNode:
    def __init__(self, title):
        self.title = title
        self.prev = None
        self.next = None


class MusicPlayer:
    def __init__(self):
        self.head = None
        self.current = None

    # Add song at end of circular doubly linked list
    def add_song(self, title):
        new_node = SongNode(title)

        if self.head is None:
            self.head = new_node
            new_node.next = new_node
            new_node.prev = new_node
            self.current = new_node
        else:
            last = self.head.prev

            last.next = new_node
            new_node.prev = last

            new_node.next = self.head
            self.head.prev = new_node

    def play_current(self):
        if self.current:
            print(f"Now Playing: {self.current.title}")
        else:
            print("Playlist is empty.")

    # Next song
    def next_song(self):
        if self.current:
            self.current = self.current.next
            self.play_current()

    # Previous song
    def prev_song(self):
        if self.current:
            self.current = self.current.prev
            self.play_current()

```
